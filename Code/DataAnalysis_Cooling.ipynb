```python
data = []

for month in months:
    days = days_in_month[month - 1]
    for day in range(1, days + 1):
        for hour in hours:
            if hour in Off_peak_hours:
                energy_demand = 0
            else:
                energy_demand = Energy_Demand
            data.append({'Day': day, 'Month': month, 'Hour': hour, 'Energy Demand (MW)': energy_demand})

df_Energy_Demand = pd.DataFrame(data)
```

```python
#Create dictionary with ASIC data

column_pairs = ['Release', 'Hashrate (Th/s)', 'Power (W)', 'Algorithm',
       'Income ($/day)', 'Electricity Cost ($/day)', 'With Electricity',
       'Without Electricity', 'Efficiency (j/Gh)', 'Price ($)', 'Noise db',
       'Cooling', 'Seller', 'Days to Breakeven']
dicts = {}

for col in column_pairs:
    dict_name = f"dict_{col}"
    dicts[dict_name] = ASIC_df.set_index('Model')[col].to_dict()

for name, dictionary in dicts.items():
    print(f"{name} -> {dictionary}")
```

```python
def calculate_rigs_per_model(df, power_dict, energy_column):

    for model, power in power_dict.items():
        df[model] = (df[energy_column] * 1_000_000 / power).round().astype(int)
    return df
energy_column = 'Energy for Mining (MW)'
df_10 = calculate_rigs_per_model(df_10,dicts['dict_Power (W)'],energy_column)
df_20 = calculate_rigs_per_model(df_20,dicts['dict_Power (W)'],energy_column)
df_30 = calculate_rigs_per_model(df_30,dicts['dict_Power (W)'],energy_column)
df_40 = calculate_rigs_per_model(df_40,dicts['dict_Power (W)'],energy_column)
df_50 = calculate_rigs_per_model(df_50,dicts['dict_Power (W)'],energy_column)
```

```python
columns_to_exclude = ['Date', 'Hour', 'Energy for Mining (MW)',
       #'MicroBT WhatsMiner M63S',
       #'Bitmain Antminer S21 Hyd (335Th)', 'MicroBT WhatsMiner M63',
       #'MicroBT WhatsMiner M66S', 'Bitmain Antminer S21 Pro (234Th)',
       #'Bitmain Antminer S21 (200Th)', 'Bitmain Antminer T21 (190Th)',
       #'MicroBT WhatsMiner M60S', 'Canaan Avalon A1566',
       #'MicroBT WhatsMiner M60', 'Bitmain Antminer S19j Pro (104Th)',
       #'Bitmain Antminer S19 Pro (110Th)'
       ]
```

```python
def calculate_max_values(df, columns_to_exclude):

    numeric_columns = df.drop(columns=columns_to_exclude).select_dtypes(include=[np.number])
    max_values = numeric_columns.max().to_dict()
    return max_values


max_values_10 = calculate_max_values(df_10, columns_to_exclude)
max_values_20 = calculate_max_values(df_20, columns_to_exclude)
max_values_30 = calculate_max_values(df_30, columns_to_exclude)
max_values_40 = calculate_max_values(df_40, columns_to_exclude)
max_values_50 = calculate_max_values(df_50, columns_to_exclude)
```

```python
def plot_max_values(max_values, title):

    plt.figure(figsize=(12, 8))
    bars = sns.barplot(x=list(max_values.keys()), y=list(max_values.values()), palette="viridis")
    
    plt.title(title, fontsize=16)
    plt.xticks(rotation=90)
    plt.ylabel('Maximum Value')
    plt.xlabel('Model')
    plt.yticks([])

    # Add labels to each bar
    for bar in bars.patches:
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height(),
            f'{int(bar.get_height())}',
            ha='center', 
            va='bottom'
        )

    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
plot_max_values(max_values_10, title='Maximum Number of Rigs for Each ASIC Model at 10%')
plot_max_values(max_values_20, title='Maximum Number of Rigs for Each ASIC Model at 20%')
plot_max_values(max_values_30, title='Maximum Number of Rigs for Each ASIC Model at 30%')
plot_max_values(max_values_40, title='Maximum Number of Rigs for Each ASIC Model at 40%')
plot_max_values(max_values_50, title='Maximum Number of Rigs for Each ASIC Model at 50%')
```

```python
def plot_combined_max_values(max_values_dict, title="Number of Rigs per Cooling Rate"):

    # Convert max_values_dict to a DataFrame for easier plotting
    df = pd.DataFrame(max_values_dict)

    # Plot
    plt.figure(figsize=(14, 8))
    df.plot(kind="bar", ax=plt.gca(), colormap="viridis", width=0.8)

    # Set title, labels, and rotate x-axis labels
    plt.title(title, fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.ylabel("Maximum Value")
    plt.xlabel("Model")

    # Annotate each bar with its height
    for container in plt.gca().containers:
        plt.bar_label(container, fmt='%d', label_type='edge', padding=3)

    plt.legend(title="Percentage")
    plt.tight_layout()
    plt.show()

# Example usage
max_values_dict = {
    '10%': max_values_10,
    '20%': max_values_20,
    '30%': max_values_30,
    '40%': max_values_40,
    '50%': max_values_50
}

plot_combined_max_values(max_values_dict, title="Number of Rigs per Cooling Rate")
```
